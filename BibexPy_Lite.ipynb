## 2B. Use your own data

Set a project name, then upload your **Web of Science `.txt`** and **Scopus `.csv`** exports (one or more of each).

```python
import os
from google.colab import files

project = 'My Project'  # <-- change to your project name
data_dir = f'Workspace/{project}/Data'
os.makedirs(data_dir, exist_ok=True)

print('Upload your WoS (.txt) and Scopus (.csv) files:')
uploaded = files.upload()
for name in list(uploaded):
    os.replace(name, os.path.join(data_dir, name))
print('Saved to', data_dir, '->', os.listdir(data_dir))
```

```python
from merge import run_project
out = run_project(f'Workspace/{project}')
print('Results saved to:', out)
```

## 3. Download the results

Zips the analysis folder (Merged.xlsx, Merged_Vos.txt, Borderline, etc.) and downloads it.

```python
import shutil
from google.colab import files
zip_path = shutil.make_archive(out, 'zip', out)
files.download(zip_path)
```
