## 2. (Optional) Add your own data

Skip this to try the bundled **Sample Project**. To merge your own data, run this cell: enter a project name, then upload your **WoS `.txt`** and **Scopus `.csv`** exports (one or more of each).

```python
import os
from google.colab import files

project = input('Project name (e.g. My Review): ').strip() or 'My Project'
data_dir = f'Workspace/{project}/Data'
os.makedirs(data_dir, exist_ok=True)

print(f'Upload WoS (.txt) and Scopus (.csv) files for "{project}":')
for name in list(files.upload()):
    os.replace(name, os.path.join(data_dir, name))
print('Saved to', data_dir, '->', os.listdir(data_dir))
```

## 4. Download the results

Results are saved under `Workspace/<project>/Analysis_<timestamp>/` (`Merged.xlsx`, `Merged_Vos.txt`, `Borderline_Uncertain.xlsx`, ...). Download them from the **Files** panel on the left, or run this cell to zip & download the most recent run.

```python
import glob, os, shutil
from google.colab import files

runs = sorted(glob.glob('Workspace/*/Analysis_*'), key=os.path.getmtime)
assert runs, 'No analysis folder found yet — run cell 3 first.'
latest = runs[-1]
print('Zipping:', latest)
files.download(shutil.make_archive(latest, 'zip', latest))
```
